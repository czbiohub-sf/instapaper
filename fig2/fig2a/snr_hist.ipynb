```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

a = 0.4
#colors = [(0, 0, 204/255, a), (0, 0, 255/255, a), (51/255, 102/255, 255/255, a), (102/255, 102/255, 255/255, a), (153/255, 51/255, 255/255, a), (204/255, 51/255, 255/255, a), (255/255, 102/255, 255/255, a)]
colors = [(0, 153/255, 255/255, a), (102/255, 0, 255/255, a), (255/255, 0, 0, a), (230/255, 115/255, 0/255, a), (255/255, 204/255, 0/255, a), (0, 204/255, 0/255, a), (0, 153/255, 153/255, a)]

fig = plt.figure(figsize=(14,8))
handles = []
for snr_mu, color in zip(range(5, 13), colors):
    csv_filename = 'snr_'+str(snr_mu)+'_1_numspots_50_coord_snr_list.csv'
    csv_filepath = os.getcwd() + '/v2/spot_data/' + csv_filename
    df = pd.read_csv(csv_filepath)
    snr_vals = list(np.asarray(df)[:,2])
    plt.hist(snr_vals, fc=color, bins = np.arange(min(snr_vals), max(snr_vals)+1, 0.5)-0.5)
    handles.append(mpatches.Patch(color=color, label='Mean SNR of image = ' + str(snr_mu)))
plt.legend(handles = handles)
plt.xticks(range(5,13))
plt.xlabel('Frequency')
plt.ylabel('SNR')
plt.title('SNR per spot')
plt.show()
```
