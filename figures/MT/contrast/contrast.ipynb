```python
from skimage.io import imread
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def f(x, mean):
    return (x-mean)**2
```

```python
for a in range(0,4):
    for b in range(0,4):
        filename = ('ISS_rnd%s_ch%s_z0.png' % (a,b))
        print(filename)
        values = imread(filename)
        RMSD_values = f(values, np.mean(values))
        plt.figure()
        plt.hist(RMSD_values)
        plt.xlim(0, 0.4*1e9)
        plt.show()
```
