```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
```

```python
ISS_rnd0_ch1_z0_results = (.95, .70)
ISS_rnd0_ch3_z0_results = (.92, .89)
ISS_rnd1_ch1_z0_results = (.81, .76)
```

```python
ind = np.arange(2)
width = 0.35 
tick_loc = ind + width
bar_width = 0.3

plt.figure(figsize=(8,5))
plt.bar(tick_loc-bar_width, ISS_rnd0_ch1_z0_results, bar_width, label='ISS_rnd0_ch1_z0_results', color='#d7b5d8')
plt.bar(tick_loc, ISS_rnd0_ch3_z0_results, bar_width, label='ISS_rnd0_ch3_z0_results', color='#41b6c4')
plt.bar(tick_loc+bar_width, ISS_rnd1_ch1_z0_results, bar_width, label='ISS_rnd1_ch1_z0_results', color='#a1dab4')

plt.ylim(0, 1.3)
plt.xticks(tick_loc, ('Precision', 'Recall'))
plt.yticks(np.arange(0, 1.1, 0.2))
plt.legend(loc='upper right', prop={'size': 12})
plt.savefig('rca_results_bar_alternate.pdf', transparent=True, bbox_inches="tight")
plt.show()
```

```python
precision = [.95, .92, .81]
recall = [.70, .89, .76]
```

```python
x = ['Precision', 'Recall']
mean = [np.mean(precision), np.mean(recall)]
std = [np.std(precision), np.std(recall)]

plt.figure(figsize=(8,3))

ind = np.arange(2)
width = 0.2

x_pos = [i for i, _ in enumerate(x)]

plt.bar(x_pos, mean, color=['#b2df8a', '#1f78b4'], yerr=std)
plt.xticks(ind + width/2, ('Precision', 'Recall'))

leg_elem_1 = Line2D([0], [0], marker='o', color='black', markeredgecolor='None', markerfacecolor='None', label='stdev', markersize=10)
plt.legend(handles=[leg_elem_1], loc='upper center', frameon=1, prop={'size': 15}, framealpha=0.5)


plt.ylim(0, 1.1)
plt.xticks(x_pos, x)

plt.savefig('rca_results_bar.pdf', transparent=True, bbox_inches="tight")

plt.show()
```
