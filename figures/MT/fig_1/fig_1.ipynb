```python
from skimage.io import imread
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
im = imread('sample_image_unfiltered.png', as_gray=True)
plt.figure(figsize=(13.9,10.4))
plt.imshow(im, cmap='gray')
plt.xticks([])
plt.yticks([])
plt.xlim(550,850)
plt.ylim(400,700)
plt.savefig("sample_image_unfiltered.pdf", transparent=True, bbox_inches="tight")
```

```python
im = imread('sample_image_filtered.png', as_gray=True)
plt.figure(figsize=(13.9,10.4))
plt.imshow(im, cmap='gray')
plt.xticks([])
plt.yticks([])
plt.xlim(550,850)
plt.ylim(400,700)
plt.savefig("sample_image_filtered.pdf", transparent=True, bbox_inches="tight")
```

```python
im = imread('sample_image.png', as_gray=True)
plt.figure(figsize=(13.9,10.4))
ref = pd.read_csv('sample_expert_annotations.csv')
ref = np.asarray(ref)
for x, y in ref:
    plt.scatter([x-1], [y-1], color='y', s=200)
plt.xticks([])
plt.yticks([])
plt.xlim(550,850)
plt.ylim(400,700)
plt.imshow(im, cmap='gray')
plt.savefig("sample_image_expert_annotations.pdf", transparent=True, bbox_inches="tight")
```
