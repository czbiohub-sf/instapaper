```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
wo_cropping = (.47, .24)
w_cropping = (.97, .88)
ind = np.arange(2)
width = 0.35 
plt.figure(figsize=(8,5))
plt.bar(ind, wo_cropping, width, label='without cropping', color='xkcd:goldenrod')
plt.bar(ind + width, w_cropping, width, label='with cropping', color='xkcd:indigo')
plt.title('Effect of Autocropping on Precision and Recall')
plt.ylim(0, 1.1)
plt.xticks(ind + width/2, ('Precision', 'Recall'))
plt.legend(loc='upper right', prop={'size': 12})
plt.savefig('autocropping_effect_bar.pdf', transparent=True, bbox_inches="tight")
plt.show()
```

```python
precision = (0.47, 0.97)
recall = (0.24, 0.88)

plt.figure(figsize=(3,4))
plt.bar(x=[1,1.75], height=precision, color=['xkcd:light gray', 'xkcd:gray'], edgecolor=['black', 'black'])
plt.xticks([1,1.75], ['w/out cropping', 'w/ cropping'])
plt.yticks([0, 0.5, 1])
plt.ylabel('Precision')
plt.savefig('autocropping_effect_bar_precision.pdf', transparent=True, bbox_inches="tight")
plt.show()

plt.figure(figsize=(3,4))
plt.bar(x=[1,1.75], height=recall, color=['xkcd:light gray', 'xkcd:gray'], edgecolor=['black', 'black'])
plt.xticks([1,1.75], ['w/out cropping', 'w/ cropping'])
plt.yticks([0, 0.5, 1])
plt.ylabel('Recall')
plt.savefig('autocropping_effect_bar_recall.pdf', transparent=True, bbox_inches="tight")
plt.show()
```
