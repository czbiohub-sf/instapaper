```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
precision = np.asarray(pd.read_csv('precision.csv'))[:,1:]
precision = np.where(precision == 'None', 0, precision).astype(np.float)

recall = np.asarray(pd.read_csv('recall.csv'))[:,1:]
recall = np.where(recall == 'None', 0, recall).astype(np.float)

num_replicates_list = np.arange(2,25,1)
num_spots_list = np.arange(50,250,25)
```

```python
cm = 'plasma_r'
```

```python
n_replicates, n_spots, precisions, recalls = [], [], [], []

for i, num_replicates in enumerate(num_replicates_list[:-1]):
    for j, num_spots in enumerate(num_spots_list):
        n_replicates.append(num_replicates)
        n_spots.append(num_spots)
        precisions.append(precision[i][j])
        recalls.append(recall[i][j])
```

```python
plt.figure()
plt.scatter(n_spots, n_replicates, c=precisions, cmap=cm, s=80)
plt.colorbar()
plt.clim(0,1)
plt.xticks(num_spots_list)
plt.title('Average Precision')
plt.xlabel('Number of Spots in the Image')
plt.ylabel('Number of Workers')
plt.savefig('worker_redundancy_precision.pdf', transparent=True, bbox_inches='tight')
plt.show()
```

```python
plt.figure()
plt.scatter(n_spots, n_replicates, c=recalls, cmap=cm, s=80)
plt.colorbar()
plt.clim(0,1)
plt.xticks(num_spots_list)
plt.title('Average Recall')
plt.xlabel('Number of Spots in the Image')
plt.ylabel('Number of Workers')
plt.savefig('worker_redundancy_recall.pdf', transparent=True, bbox_inches='tight')
plt.show()
```
