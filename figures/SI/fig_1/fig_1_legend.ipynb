```python
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
```

```python
plt.figure(figsize=(12,14))
h1 = Line2D([0],[0], marker='o', color='None', markerfacecolor='None', markeredgecolor='#f88379', label='clumpy cluster', markersize=60)
h2 = Line2D([0],[0], marker='o', color='None', markerfacecolor='None', markeredgecolor='xkcd:green', label='declumped cluster 1', markersize=60)
h3 = Line2D([0],[0], marker='o', color='None', markerfacecolor='None', markeredgecolor='xkcd:violet', label='declumped cluster 2', markersize=60)
h4 = Line2D([0],[0], marker='*', color='None', markerfacecolor='yellow', markeredgecolor='black', label='expert annotation', markersize=60)
h5 = Line2D([0],[0], marker='o', color='None', markerfacecolor='xkcd:aqua', markeredgecolor='None', label='worker annotation', markersize=26)
h6 = Line2D([0],[0], marker='o', color='None', markerfacecolor='xkcd:lilac', markeredgecolor='None', label='worker annotation', markersize=26)

legend = plt.legend(handles=[h1, h2, h3, h4, h5, h6], loc='upper right', frameon=0, prop={'size': 50})
frame = legend.get_frame()
frame.set_facecolor('white')
plt.xticks([])
plt.yticks([])
plt.axis('off')

plt.savefig("legend", transparent=True, bbox_inches="tight")

plt.show()
```
